```python
import pandas as pd

df = pd.read_csv('Car_Details_DataSet.csv')
```

## **Operations on Car Details DataSet**

Filtering Cars by Transmission Type

```python
automatic_cars = df[df['transmission'] == 'Automatic']
automatic_cars.head()
```

 Calculate the Average Selling Price by Fuel Type

```python
# Group by fuel type and calculate average selling price
average_price_by_fuel = df.groupby('fuel')['selling_price'].mean()
print(average_price_by_fuel)
```

Count the Number of Cars by Owner Type

```python
# Count the number of cars by owner type
cars_by_owner = df['owner'].value_counts()
print(cars_by_owner)
```

Calculate the Average Kilometers Driven by Seller Type

```python
# Group by seller type and calculate average kilometers driven
average_km_by_seller = df.groupby('seller_type')['km_driven'].mean()
print(average_km_by_seller)
```
