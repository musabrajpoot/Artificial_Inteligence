```python
import pandas as pd # Import the pandas library

# Assuming your data is in a CSV file named 'vegetable_data.csv'
df = pd.read_csv('laptop_price_dataset.csv')
```

Sorting the laptop price in Desending Order (euro).

```python
# Sorting laptops by 'Price (Euro)' in descending order
sorted_laptops = df.sort_values(by='Price (Euro)', ascending=False)
sorted_laptops.head()  # Show top sorted rows
```

Calculate the Average Price of Laptops


```python
# Calculate the average price of all laptops
average_price = df['Price (Euro)'].mean()
print(f'The average price of laptops is: {average_price:.2f} Euros')
```

Count the Number of Laptops from Each Company


```python
# Group by 'Company' and count the number of laptops
laptop_counts = df['Company'].value_counts()
print(laptop_counts)
```

Find the Laptop with the Maximum Price

```python
# Find the laptop with the maximum price
max_price_laptop = df.loc[df['Price (Euro)'].idxmax()]
print('The most expensive laptop is:')
print(max_price_laptop)
```

Find the Cheapest Laptop


```python
# Find the laptop with the lowest price
cheapest_laptop = df.loc[df['Price (Euro)'].idxmin()]
print('The cheapest laptop is:')
print(cheapest_laptop)
```

Count the Number of Laptops by Screen Resolution

```python
# Count the number of laptops by screen size
screen_size_counts = df['ScreenResolution'].value_counts()
print(screen_size_counts)
```

Find Laptops with SSD or HDD

```python
# Filter laptops with SSD storage
ssd_laptops = df[df['Memory'].str.contains('SSD', case=False)]
ssd_laptops.head()

```

```python
# Filter laptops with HDD storage
hdd_laptops = df[df['Memory'].str.contains('HDD', case=False)]
hdd_laptops.head()
```

Filter Laptops by Processor Type

```python
intel_laptops = df[df['GPU_Company'].str.contains('Intel', case=False)]
intel_laptops.head()
```

```python
amd_laptops = df[df['GPU_Company'].str.contains('AMD', case=False)]
amd_laptops.head()
```
