```python
import pandas as pd
df = pd.read_csv('ElectricCar_DataSet.csv')
```

## **Operations on Electric Cars DataSet**

Count Unique Values for a Specific Column

```python
# Count unique values in the 'Brand' column
df['Brand'].value_counts()
```

Filtering Data

```python
# Filter rows where TopSpeed_KmH is greater than 200
fast_ev = df[df['TopSpeed_KmH'] > 200]
fast_ev.head()
```

Finding Max/Min Values

```python
# EV with the maximum range
max_range_ev = df[df['Range_Km'] == df['Range_Km'].max()]
max_range_ev

```
