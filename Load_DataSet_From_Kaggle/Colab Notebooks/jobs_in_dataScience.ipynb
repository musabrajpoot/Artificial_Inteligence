```python
import pandas as pd

df = pd.read_csv('jobs_in_dataScience.csv')
```

## **Operations on Jobs in DataScience **

Filtering Data Based on Job Titles

```python
# Filter data for specific job title, e.g., "Data Scientist"
data_scientist_jobs = df[df['job_title'] == 'Data Scientist']

# Show some basic info about these jobs
print(data_scientist_jobs[['job_title', 'salary_in_usd', 'experience_level', 'company_location']].head())
```

Analyzing Job Categories by Salary

```python
# Group by job category and calculate the average salary in USD
avg_salary_by_category = df.groupby('job_category')['salary_in_usd'].mean()

# Display the result
print(avg_salary_by_category)
```

 Visualizing Experience Level and Salary Trends

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Group by experience level and calculate the average salary
avg_salary_by_experience = df.groupby('experience_level')['salary_in_usd'].mean().reset_index()

# Plot the average salary by experience level
sns.barplot(x='experience_level', y='salary_in_usd', data=avg_salary_by_experience)
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary (USD)')
plt.show()
```

 Analyzing Salary Differences by Employment Type

```python
# Group by employment type and calculate the average salary in USD
avg_salary_by_employment_type = df.groupby('employment_type')['salary_in_usd'].mean()

# Display the result
print(avg_salary_by_employment_type)

# Visualize salary differences by employment type
sns.barplot(x='employment_type', y='salary_in_usd', data=df)
plt.title('Average Salary by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Salary (USD)')
plt.show()
```
