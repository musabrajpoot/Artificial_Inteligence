```python
import pandas as pd

df = pd.read_csv('Software_Engineer_Salaries.csv')
```

## Operationson Software_Engineer_Salaries DataSet

Sort Jobs by Salary


```python
# Sort jobs by 'Salary' in descending order
sorted_jobs_by_salary = df.sort_values(by='Salary', ascending=False)
sorted_jobs_by_salary.head()  # Show top sorted rows
```

Filter Jobs by Location

```python
# Filter jobs by location (e.g., New York)
jobs_in_location = df[df['Location'].str.contains('New York', case=False , na=False)]
jobs_in_location.head()  # Show filtered jobs
```

Calculate the Average Salary

```python
# Calculate the average salary
# Convert 'Salary' column to numeric, handling errors
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')

average_salary = df['Salary'].mean()
print(f'The average salary is: {average_salary:.2f}')
```

Count the Number of Jobs from Each Company

```python
# Group by 'Company' and count the number of job listings
job_counts_by_company = df['Company'].value_counts()
print(job_counts_by_company)
```

Filter Jobs by Minimum Company Score

```python
# Filter jobs with a company score of at least 4.0
high_score_jobs = df[df['Company Score'] >= 4.0]
high_score_jobs.head()  # Show jobs with high company scores
```
