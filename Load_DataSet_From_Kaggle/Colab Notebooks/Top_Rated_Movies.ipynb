```python
import pandas as pd

df = pd.read_csv('Top_Rated_Movies.csv')
```

## **Operations on Movies DataSet**

Top 5 Most Popular Movies

```python
# Sort the dataset by 'popularity' and get the top 5 movies
top_5_popular_movies = df[['title', 'popularity']].sort_values(by='popularity', ascending=False).head(5)
print(top_5_popular_movies)
```

Movies Released After a Specific Year

```python
# Convert 'release_date' to datetime format and filter movies released after 2000
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
movies_after_2000 = df[df['release_date'].dt.year > 2000]
print(movies_after_2000[['title', 'release_date']])
```

Average Vote Score of All Movies

```python
# Calculate the average of 'vote_average'
average_vote_score = df['vote_average'].mean()
print(f'Average vote score: {average_vote_score:.2f}')
```

Highest Rated Movie

```python
# Find the movie with the highest 'vote_average'
highest_rated_movie = df[['title', 'vote_average']].sort_values(by='vote_average', ascending=False).head(1)
print(highest_rated_movie)
```

Movies with Vote Average Above 8

```python
# Filter movies with 'vote_average' greater than 8
high_rated_movies = df[df['vote_average'] > 8]
print(high_rated_movies[['title', 'vote_average']])
```
