```python
import pandas as pd

df = pd.read_csv('climate_change_impact_on_agriculture_2024.csv')
```

# Operation on climate_change_impact_on_agriculture_2024 **DataSet**

Calculate Average Crop Yield

```python
# Calculate the average crop yield
average_crop_yield = df['Crop_Yield_MT_per_HA'].mean()
print(f'The average crop yield is: {average_crop_yield:.2f} MT/HA')
```

Filter Data by Region and Crop Type

```python
# Filter data for a specific region and crop type (e.g., Region: 'Asia', Crop Type: 'Wheat')
filtered_data = df[(df['Region'] == 'North') & (df['Crop_Type'] == 'Wheat')]
filtered_data.head()
```

Filter Countries with High Irrigation Access

```python
# Filter countries with irrigation access above 75%
high_irrigation_countries = df[df['Irrigation_Access_%'] > 75]
high_irrigation_countries.head()
```

Average Temperature by Region

```python
# Group by 'Region' and calculate the average temperature
average_temp_by_region = df.groupby('Region')['Average_Temperature_C'].mean()
print(average_temp_by_region)
```
