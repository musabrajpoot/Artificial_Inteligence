```python
import requests # correct the typo to lowercase 'r'
from bs4 import BeautifulSoup
```

```python
Response = requests.get("http://www.sahiwal.comsats.edu.pk/") # correct the typo to lowercase 'r'
```

```python
Content = Response.content # correct the typo to lowercase 'r'
```

```python
print(Content) # correct the typo to lowercase 'r'
```

```python
soup = BeautifulSoup(Content, "html.parser") # correct the typo to lowercase 'r'
for i in soup.find_all('a'): # correct the typo to lowercase 'r'
    print(i.get_text('href')) # correct the typo to lowercase 'r'
```
