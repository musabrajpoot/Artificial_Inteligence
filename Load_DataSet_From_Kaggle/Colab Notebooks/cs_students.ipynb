```python
import pandas as pd

df = pd.read_csv('cs_students.csv')
```

# **Operation on CS_Students Career Predition**

Average GPA by Major

```python
# Group by 'Major' and calculate the average GPA
average_gpa_by_major = df.groupby('Major')['GPA'].mean()
print(average_gpa_by_major)
```

Top 5 Students with Highest GPA

```python
# Sort by 'GPA' in descending order and get top 5 students
top_5_students_gpa = df[['Name', 'GPA']].sort_values(by='GPA', ascending=False).head(5)
print(top_5_students_gpa)
```

Count of Students by Interested Domain

```python
# Count the number of students for each interested domain
students_by_domain = df['Interested Domain'].value_counts()
print(students_by_domain)
```

Number of Projects by Major

```python
# Group by 'Major' and sum the number of projects
projects_by_major = df.groupby('Major')['Projects'].sum()
print(' ' + projects_by_major)
```

Students Interested in Machine Learning

```python
# Filter students whose interested domain includes 'Machine Learning'
ml_students = df[df['Interested Domain'].str.contains('Machine Learning', case=False)]
print(ml_students[['Name', 'GPA', 'Major']])
```
