```python
import pandas as pd

df = pd.read_csv('Customer_Purchasing_Behaviors.csv')
```

# **Operations on Customer_Purchasing_Behaviors DataSet**

Calculate Average Purchase Amount

```python
# Calculate the average purchase amount
average_purchase_amount = df['purchase_amount'].mean()
print(f'The average purchase amount is: {average_purchase_amount:.2f}')
```

Sort Users by Loyalty Score

```python
# Sort users by loyalty score in descending order
sorted_by_loyalty = df.sort_values(by='loyalty_score', ascending=False)
sorted_by_loyalty.head()
```

Filter Users by Region

```python
# Filter users by region (e.g., 'North')
users_in_region = df[df['region'].str.contains('North', case=False)]
users_in_region.head()
```

Correlation between Annual Income and Purchase Amount

```python
# Calculate the correlation between 'annual_income' and 'purchase_amount'
correlation_income_purchase = df['annual_income'].corr(df['purchase_amount'])
print(f'The correlation between annual income and purchase amount is: {correlation_income_purchase:.2f}')
```

Find Users with High Annual Income

```python
# Filter users with annual income greater than 100,000
high_income_users = df[df['annual_income'] > 60000]
high_income_users.head()
```

Find the User with the Maximum Purchase Amount

```python
# Find the user with the maximum purchase amount
max_purchase_user = df.loc[df['purchase_amount'].idxmax()]
print(f'The user with the maximum purchase amount is:\n{max_purchase_user}')
```
