```python
import pandas as pd
df = pd.read_csv('WWE_Champion_DataSet.csv')
```

## **Operations on WWE Champion DataSet**

Unique Events and Matches

```python
# Get unique values in the 'Event' column
df['Event'].unique()

# Get unique values in the 'Title Match' column
df['Title Match'].unique()
```

Count the Number of Matches per Event

```python
# Count the number of matches for each event
df['Event'].value_counts()
```

Filtering Data

```python
# Filter only Title Matches
title_matches = df[df['Title Match'] == 'Yes']
title_matches.head()
```

Find the Most Frequent Winner

```python
# Find the wrestler with the most wins
most_wins = df['Winner'].value_counts().head(1)
print("Wrestler with most wins:", most_wins)
```

Find the Most Frequent Loser

```python
# Find the wrestler with the most wins
most_wins = df['Loser'].value_counts().head(1)
print("Wrestler with most wins:", most_wins)
```
