```python
import pandas as pd
df = pd.read_csv('ipl2024_Matches_DataSet.csv')
```

## **Operations on IPL2024 Matches DataSet**

Filter by Specific Columns

```python
# Filter the DataFrame by selecting specific columns
df_filtered = df[['team1', 'team2', 'winner', 'player_of_the_match']]
df_filtered.head()
```

Count Matches Played by Each Team

```python
# Count the number of matches played by each team
team_matches = pd.concat([df['team1'], df['team2']]).value_counts()
print(team_matches)
```

Filter Matches Based on the Toss Decision

```python
# Filter matches where the toss decision was to bat
bat_decision = df[df['decision'] == 'Bat']
bat_decision.head()

# Filter matches where the toss decision was to field
field_decision = df[df['decision'] == 'Field']
field_decision.head()
```

Count Wins by Each Team

```python
# Count the number of matches won by each team
wins_by_team = df['winner'].value_counts()
print(wins_by_team)
```

Find Matches Won by Toss Winners

```python
# Filter matches where the toss winner was also the match winner
toss_and_match_winner = df[df['toss_winner'] == df['winner']]
print(f"Matches where the toss winner won the match: {len(toss_and_match_winner)}")
```
