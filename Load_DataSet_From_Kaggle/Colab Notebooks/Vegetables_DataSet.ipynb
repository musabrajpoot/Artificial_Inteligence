```python
import pandas as pd

# Load the dataset (use the correct path if uploaded through Google Drive)
df = pd.read_csv('/content/vegetable_prices_dataset.csv')  # If you uploaded manually
# df = pd.read_csv('/content/drive/MyDrive/vegetables.csv')  # If using Google Drive

# Display the dataset
print(df)
```
