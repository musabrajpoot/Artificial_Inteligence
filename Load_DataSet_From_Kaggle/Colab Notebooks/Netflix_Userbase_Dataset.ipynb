```python
import pandas as pd

df = pd.read_csv('Netflix_Userbase_Dataset.csv')
```

## **Operations on Netflix_Userbase_Dataset**

Filtering Data

```python
# Filter users by subscription type
premium_users = df[df['Subscription Type'] == 'Premium']
premium_users.head()
```

Filter users from a specific country (e.g., "USA").

```python
# Filter users by country
usa_users = df[df['Country'] == 'United States']
usa_users.head()
```

Calculating Total Revenue by Country

```python
# Group by country and calculate total revenue
revenue_by_country = df.groupby('Country')['Monthly Revenue'].sum().reset_index()
print(revenue_by_country)
```

Calculate Average Age per Subscription Type

```python
# Group by subscription type and calculate average age
average_age_by_subscription = df.groupby('Subscription Type')['Age'].mean()
print(average_age_by_subscription)
```
