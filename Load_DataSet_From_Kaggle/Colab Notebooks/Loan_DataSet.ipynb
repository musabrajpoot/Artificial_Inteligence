```python
import pandas as pd
df = pd.read_csv('Loan_DataSet.csv')
```

## **Operations on Loan DataSet**

View the First/Last Few Rows

```python
df.head()

df.tail()
```

Value Counts for a Specific Column

```python
# Count unique values in the 'Loan_Status' column
df['Loan_Status'].value_counts()

# Count unique values in the 'Gender' column
df['Gender'].value_counts()
```

Filtering Rows

```python
# Filter rows where ApplicantIncome is greater than 5000
high_income = df[df['ApplicantIncome'] > 5000]
high_income.head()
```

Sorting Data

```python
# Sort by LoanAmount in ascending order
df.sort_values(by='LoanAmount', inplace=True)

# Sort by LoanAmount in descending order
df.sort_values(by='LoanAmount', ascending=False, inplace=True)

print(df.sort_values)
```
