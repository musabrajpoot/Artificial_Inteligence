```python
import pandas as pd
df = pd.read_csv('spotify-2023.csv',  encoding='latin-1')
```

## **Operation on Spotify DataSet**

Top 10 Most Streamed Songs

```python
# Sort the dataset by 'streams' and get the top 10 songs
top_10_streamed_songs = df[['track_name', 'artist(s)_name', 'streams']].sort_values(by='streams', ascending=False).head(10)
print(top_10_streamed_songs)
```

Filter Songs Released After a Specific Year

```python
# Filter songs released after 2015
songs_after_2015 = df[df['released_year'] > 2015]
print(songs_after_2015[['track_name', 'artist(s)_name', 'released_year']])
```

Average BPM of All Songs

```python
# Calculate the average BPM of the songs
average_bpm = df['bpm'].mean()
print(f'Average BPM: {average_bpm:.2f}')
```

Top 5 Most Danceable Songs

```python
# Sort the dataset by 'danceability_%' and get the top 5 songs
top_5_danceable_songs = df[['track_name', 'artist(s)_name', 'danceability_%']].sort_values(by='danceability_%', ascending=False).head(5)
print(top_5_danceable_songs)
```
