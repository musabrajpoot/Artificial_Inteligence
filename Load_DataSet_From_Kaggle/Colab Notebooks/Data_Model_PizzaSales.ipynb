```python
import pandas as pd

df = pd.read_excel('Data_Model_PizzaSales.xlsx')
```

## **Operations on Piza Sales DataSet**

Calculate The Total Revenue

```python
total_revenue = df['total_price'].sum()
print(f"Total Revenue: {total_revenue}")
```

Calculate the average revenue per order

```python
avg_revenue_per_order = df['total_price'].mean()
print(f"Average Revenue per Order: {avg_revenue_per_order}")
```

```python
# Plot the revenue trend over time
import matplotlib.pyplot as plt
import seaborn as sns
```

Analyzing Pizza Orders by Size

```python
# Count the number of orders by pizza size
pizza_size_count = df['pizza_size'].value_counts()

# Plot the number of orders by pizza size
pizza_size_count.plot(kind='bar', title='Number of Orders by Pizza Size')
plt.xlabel('Pizza Size')
plt.ylabel('Number of Orders')
plt.show()
```

Finding the Most Popular Pizza Categories

```python
# Count the number of orders by pizza category
pizza_category_count = df['pizza_category'].value_counts()

# Plot the number of orders by pizza category
pizza_category_count.plot(kind='bar', title='Number of Orders by Pizza Category')
plt.xlabel('Pizza Category')
plt.ylabel('Number of Orders')
plt.show()
```
