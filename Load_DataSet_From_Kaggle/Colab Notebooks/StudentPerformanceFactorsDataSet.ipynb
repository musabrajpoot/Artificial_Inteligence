```python
import pandas as pd

df = pd.read_csv('StudentPerformanceFactorsDataSet.csv')
```

# **Operationson Software_Engineer_Salaries DataSet**

Calculate Average Exam Score

```python
# Calculate the average exam score
average_exam_score = df['Exam_Score'].mean()
print(f'The average exam score is: {average_exam_score:.2f}')
```

Analyze Exam Score by School Type

```python
# Group by 'School_Type' and calculate the average exam score
avg_score_by_school_type = df.groupby('School_Type')['Exam_Score'].mean()
print(avg_score_by_school_type)
```

Filter Students with High Attendance (e.g., above 90%)

```python
# Filter students with attendance above 90%
high_attendance_students = df[df['Attendance'] > 90]
high_attendance_students.head()  # Show filtered students
```

Analyze the Impact of Extracurricular Activities on Exam Score

```python
# Group by 'Extracurricular_Activities' and calculate the average exam score
avg_score_by_activities = df.groupby('Extracurricular_Activities')['Exam_Score'].mean()
print(avg_score_by_activities)
```

Filter Students with Learning Disabilities

```python
# Filter students with learning disabilities
students_with_disabilities = df[df['Learning_Disabilities'] == 'Yes']
students_with_disabilities.head()
```
