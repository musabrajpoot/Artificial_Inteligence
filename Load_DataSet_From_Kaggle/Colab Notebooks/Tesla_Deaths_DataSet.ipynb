```python
import pandas as pd
df = pd.read_csv('Tesla_Deaths_DataSet.csv')
```

### **Operations in Tesla Death DataSets**

View Summary Statistics

```python
# Get summary statistics for numerical columns
df.describe()

# Get summary statistics for all columns
df.describe(include='all')
```

Count the Number of Incidents per Year

```python
# Count the number of incidents per year
incidents_per_year = df['Year'].value_counts()
print(incidents_per_year)
```

Filter Incidents Based on the Number of Deaths

```python
# Filter incidents with more than 2 deaths
high_death_cases = df[df[' Deaths '] > 2]
high_death_cases.head()
```

Group Incidents by Country

```python
# Group by 'Country' and count the number of incidents
incidents_by_country = df.groupby(' Country ')['Case #'].count()
print(incidents_by_country)
```

Sorting Incidents by Number of Deaths

```python
# Sort incidents by the number of deaths in descending order
sorted_by_deaths = df.sort_values(by=' Deaths ', ascending=False)
sorted_by_deaths.head()
```
