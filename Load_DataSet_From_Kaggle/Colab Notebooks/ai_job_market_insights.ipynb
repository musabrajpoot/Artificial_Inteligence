```python
import pandas as pd
df = pd.read_csv('ai_job_market_insights.csv')
```

# **Operaations on Ai Job Market Inseight**

Average Salary by Industry

```python
# Group by 'Industry' and calculate the average salary
average_salary_by_industry = df.groupby('Industry')['Salary_USD'].mean()
print(average_salary_by_industry)
```

Top 5 High-Paying Jobs

```python
# Sort the data by 'Salary_USD' in descending order and get top 5 jobs
top_5_high_paying_jobs = df[['Job_Title', 'Salary_USD']].sort_values(by='Salary_USD', ascending=False).head(5)
print(top_5_high_paying_jobs)
```

Remote-Friendly Jobs

```python
# Filter jobs that are remote-friendly
remote_friendly_jobs = df[df['Remote_Friendly'] == 'Yes']
print(remote_friendly_jobs[['Job_Title', 'Location', 'Salary_USD']])
```

Filter Jobs by Required Skills

```python
# Filter jobs that require 'Python' as a skill
python_jobs = df[df['Required_Skills'].str.contains('Python', case=False)]
print(python_jobs[['Job_Title', 'Salary_USD']])
```

Average Salary by Company Size

```python
# Group by 'Company_Size' and calculate the average salary
avg_salary_by_company_size = df.groupby('Company_Size')['Salary_USD'].mean()
print(avg_salary_by_company_size)
```
